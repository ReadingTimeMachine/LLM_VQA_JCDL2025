```python
import pandas as pd
import numpy as np
```

```python
# Login using e.g. `huggingface-cli login` to access this dataset
df = pd.read_parquet("hf://datasets/ahmed-masry/ChartQAPro/data/test-00000-of-00001.parquet")
```

```python
df.head()
```

```python
questions = []

for i in range(len(df)):
    q = df.iloc[i]['Question']
    # if len(q) > 1:
    #     print('multi-question')
    #     import sys; sys.exit()
    for qq in q:
        questions.append(qq)
```

```python
len(questions)
```

```python
nAverage = 0
for q in questions:
    if 'mean' in q or 'median' in q or 'average' in q or 'typical' in q:
        print(q)
        print('')
        nAverage += 1
```
