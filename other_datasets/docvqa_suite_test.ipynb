### DocVQA

```python
data_dir = data_suite_dir + 'DocVQA/'
```

```python
with open(data_dir + 'train_v1.0_withQT.json','r') as f:
    dvqa = json.load(f)
```

```python
dvqa['data']
```

```python
countFigs = 0

for d in dvqa['data']:
    for qt in d['question_types']:
        if 'figure' in qt:
            print(d['image'])
            print(d['question'])
            print('')
            countFigs += 1
            break

print('There are:', countFigs, 'images tagged as figures')
```

```python
mean_words = ['mean', 'average', 'median', 'typical', 'minimum', 'maximum']

count = 0
for d in dvqa['data']:
    hasFig = False
    for qt in d['question_types']:
        if 'figure' in qt:
            hasFig = True
    if hasFig:
        for mw in mean_words:
            if mw.lower() in d['question'].lower():
                count += 1
                print(count, ':', d['question'])
                print('')
                break
```

So, it looks like only 2 have something "average/max" in them and ~30 total have things like min/max in them as well.
